```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ω = 2*np.pi/(24*60*60)
ϕ = np.linspace(-np.pi/2, np.pi/2, 100)
def f(ϕ):
    return 2*Ω*np.sin(ϕ)
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,5))
plt.plot(ϕ, f(ϕ), lw=3, alpha=0.8, label=r"$f(\phi)$ (Coriolis frequency)")
l = plt.plot(ϕ, ϕ*(2*Ω), lw=1.5, alpha=0.9, ls="dashed", label=r"$f(\phi) \approx \phi \frac{\partial f}{\partial \phi}|_{\phi = 0}$ (Equatorial $\beta$-plane)")
plt.plot([0], [0], "s", color=l[0].get_color())
ϕ0 = -np.pi/6
l = plt.plot(ϕ, (ϕ-ϕ0)*(2*Ω)*np.cos(ϕ0) + f(ϕ0), lw=1.5, alpha=0.9, ls="dotted", label=r"$f(\phi) \approx f(\frac{-\pi}{6}) + (\phi-(-\frac{\pi}{6})) \frac{\partial f}{\partial \phi}|_{\phi = \frac{-\pi}{6}}$ ($\beta$-plane at $\phi=\frac{-\pi}{6}$)", color="C3")
plt.plot(ϕ0, f(ϕ0), "s", color=l[0].get_color())
plt.ylim(-2*Ω, 2*Ω)
plt.xlim(-np.pi/2, np.pi/2)
plt.xlabel("latitude [in radians]")
plt.xticks([-3*np.pi/6, -2*np.pi/6, -1*np.pi/6, 0, 1*np.pi/6, 2*np.pi/6, 3*np.pi/6], [r"$-\pi/2$", r"$-\pi/3$", r"$-\pi/6$", r"$0$", r"$\pi/6$", r"$\pi/3$",r"$\pi/2$"])
plt.ylabel("angular frequency [rad/s]")

def deg2rad(x):
    return x * np.pi / 180


def rad2deg(x):
    return x * 180 / np.pi

secax = ax.secondary_xaxis('top', functions=(rad2deg, deg2rad))
secax.set_xlabel('latitude [degrees]')
secax.set_xticks([-90, -60, -30, 0, 30, 60, 90])
plt.grid(True, alpha=0.25)
plt.legend(loc="upper left")
plt.savefig("../figures/Lecture21_Coriolis_frequency_beta_plane.png", dpi=300, bbox_inches="tight")
plt.show()
```
